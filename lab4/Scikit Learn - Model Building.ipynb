```python
#Scikit-learn, or sklearn, is a Python library that provides functions for machine learning 
#and statistical modeling. Some of the functions in sklearn include: 

#Feature extraction: Extracts features from datasets in formats like text and images 
#Pipeline: Creates a pipeline of preprocessing steps and a final estimator 
#Cross-validation: Checks the accuracy of supervised models on unseen data 
#Dimensionality reduction: Reduces the number of random variables to consider 
#Feature selection: Implements feature selection algorithms 
#Naive Bayes: Implements Naive Bayes algorithms, which are supervised learning methods 
#Binarization: Divides raw data into two categories and assigns values to them using a threshold 
#Preprocessing: Converts raw data into a clean and meaningful data set 
#Data splitting, Ensemble techniques, Statistical tools, Clustering, Regression, and Classification.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
wine_data = load_wine(as_frame=True)
wine_df = wine_data.frame

print(wine_df.head())
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train,df_test=train_test_split(wine_df,train_size=0.7,test_size=0.3,random_state=100)
```
