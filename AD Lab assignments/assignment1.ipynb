# 2. Complete the following tasks using Sets

I. Set Creation and Set Length

```python
myset = {10, 11, 10, "a", "x"}
print(len(myset))
```

II. Allocation of Different Data Types into a Set

```python
mySet = {1, "hello", 3.14, True, (1, 2)}
print(mySet)
```

III. Using set() constructor

```python
mylist = [1, 3, "a", True]
mytuple = (1, 3, "a", True)

listToSet = set(mylist)
print(listToSet)

tupleToSet = set(mytuple)
print(tupleToSet)
```

IV. Accessing Set Items

```python
for item in mySet:
    print(item)
```

V. Adding items to a Set using add() and update()

```python
mySet.add("23052229")
print(mySet)

newElements = ["x", "y"]
mySet.update(newElements)
print(mySet)
```

VI. Remove set item using remove (), pop() and discard()

```python
mySet.remove("x")
print(mySet)
```

```python
mySet.pop()
print(mySet)
```

```python
mySet.discard(1)
print(mySet)
```

VII. Use clear() and del() on a set

```python
mySet.clear()
print(mySet)
```

```python
myset = {1,2,3}
print(myset)
del(myset)
```

VIII. Set Looping - for and while

```python
mySet = {1, "hello", 3.14, True, (1, 2)}
for item in mySet:
    print(item)
```

```python
setIterator = iter(mySet)
while True:
    try:
        item = next(setIterator)
        print(item)
    except StopIteration:
        break
```


IX. Set Joining using union(), update(), intersection(), intersection_update()

```python
set1 = {1, 2, 3}
set2 = {4, 5, 6}
set3 = set1.union(set2)
print(set3)
```


X. Use difference(), difference_update(), isdisjoint() and symmetric_difference()

```python
set1 = {2,3,5}
set2 = {3,5,6}
set1.difference(set2)
```

```python
set1.difference_update(set2)
print(set1)
print(set2)
set1.isdisjoint(set2)
```

```python
set1 = {1,3,5,7,9,10}
set2 = {2,4,6,8,9}
set1.symmetric_difference(set2)
```

# 3. Complete the following tasks using Dictionaries

I. Dictionary Creation &amp; Access using Key, and Updation

```python
student = {
    "roll" : 23052229,
    "name" : "Arup",
    "job" : False
}
print(student)
```

II. Dictionary Length, Type and dict() constructor

```python
car = dict(model="RB21", brand="Redbull", year=2021)
len(car)
```

```python
print(type(car))
```

III. Dictionary Adding and Removing Items (pop(), popitem()) and Checking

```python
car["driver"] = "Max Verstappen"
popped = car.pop("brand")
print(popped)
print(car)
```

```python
popped = student.popitem()
print(popped)
print(student)
```

```python
print("name" in student)
print("age" in student)
```

IV. Using clear() and del() on Dictionary

```python
del student["roll"]
print(student)
```

```python
student.clear()
print(student)
```

V. Looping and Copying Dictionary using copy()

```python
student = {
    "roll" : 23052229,
    "name" : "Arup",
    "job" : False
}
for key, value in student.items():
    print(key, value)
```

```python
studentCopy = student.copy()
print(studentCopy)
```

VI. Nested Dictionary - Creation, Accessing and Looping Items

```python
students = {
    "student1": {
        "name": "Soham",
        "age": 22,
        "course": "CE"
    },
    "student2": {
        "name": "Rahul",
        "age": 22,
        "course": "ECE"
    }
}
print(students)
```

```python
print(students["student1"])
print(students["student1"]["name"])
```

```python
students["student2"]["age"] = 23
print(students["student2"])
```
