```python
# Explore basic DataFrame operations, such as head(), tail(), info(), describe().
import pandas as pd

df = pd.read_csv('iris.csv')

print(df)
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Practice selecting and indexing data from a DataFrame.
df['species']
```

```python
df[['species', 'sepallengthcm']]
```

```python
# Filter data based on conditions.
df[df['sepallengthcm'] > 5.0]
```

```python
# Perform basic data cleaning operations, such as handling missing values.
df.fillna(df.mean(numeric_only=True))
```

```python
# Practice grouping and aggregating data.
df.groupby('species').size()
```

```python
df.groupby('species').agg(['mean', 'std'])
```

```python
# Use Pandas for basic data visualization
import matplotlib.pyplot as plt
```

```python
df['sepallengthcm'].plot(title='Sepal Length Trend')
```

```python
df['petallengthcm'].plot(kind='hist', bins=20,title='Petal Length Distribution')
```

```python
df.boxplot(column='sepalwidthcm', by='species')
```

```python
# Create a simple line plot using Matplotlib.
plt.plot(df['sepallengthcm'])
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.title("Sepal Length of Iris Dataset")

# Display the plot
plt.show()
```

```python
# Create a scatter plot using Matplotlib.
plt.scatter(df['sepallengthcm'], df['sepalwidthcm'], c='blue', label='Sepal')
plt.scatter(df['petallengthcm'], df['petalwidthcm'], c='red', label='Petal')
```

```python
# Create a bar plot using Seaborn
import seaborn as sns
sns.barplot(x='species', y='sepallengthcm', data=df)
plt.show()
```

```python
# Create a histogram using Seaborn.
sns.histplot(data=df, x='petallengthcm', bins=20, kde=True)
plt.show()
```

```python
# Create a box plot using Seaborn.
sns.boxplot(x='species', y='sepalwidthcm', data=df)
plt.show()
```

```python
# Create a heatmap using Seaborn.
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include='number')

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```
