```python
# Explore basic DataFrame operations, such as head(), tail(), info(), describe().
import pandas as pd

df = pd.read_csv('indiaCrime.csv')

print(df)
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Practice selecting and indexing data from a DataFrame.
df['value_of_property_recovered']
```

```python
df[['cases_property_stolen', 'cases_property_recovered']]
```

```python
# Filter data based on conditions.
df[df['value_of_property_recovered'] > 1000000]
```

```python
# Perform basic data cleaning operations, such as handling missing values.
df.isnull().sum()
```

```python
# Practice grouping and aggregating data.
df.groupby('area_name')['cases_property_stolen'].sum()
```

```python
df.groupby('area_name').agg({'cases_property_stolen': 'sum', 'cases_property_recovered': 'mean'})
```

```python
# Use Pandas for basic data visualization
import matplotlib.pyplot as plt
```

```python
# Property stolen over years
df.groupby('year')['value_of_property_stolen'].sum().plot()
plt.show()
```

```python
# Stolen vs recovered cases
df.plot(
    x='cases_property_stolen',
    y='cases_property_recovered',
    kind='scatter'
)
plt.show()
```

```python
# Total stolen cases by state
df.groupby('area_name')['cases_property_stolen'] \
  .sum() \
  .plot(kind='bar', figsize=(10,4))
plt.ylabel('Cases Property Stolen')
plt.title('Property Stolen Cases by State')
plt.show()
```

```python
# Compare stolen cases across years
df.boxplot(column='cases_property_stolen', by='year')
plt.title('Stolen Cases Across Years')
plt.suptitle('')
plt.show()
```
